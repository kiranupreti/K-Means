```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
a=np.array([2,3,4,7,8,9])
b=np.array([1,2,1,7,6,8])
```

```python
plt.scatter(a,b)
plt.show()
```

```python
data=np.zeros((6,2))
data[:,0]=a
data[:,1]=b
print(data.shape)
print(data)
```

```python
a=np.random.choice(range(data.shape[0]),2)#here we are choosing random element out the arrar of size 2
print(a)
                   
```

```python
#    for i in range(len(a1)):
#        centroids[i]=data[c[i],:]
#    print(centroids)
centroids= data[a,:]   
print(centroids)
centroids[0]=centroids[0,:]
centroids[1]=centroids[-1,:]

print('centroids_0 is',centroids[0])
print('centroid_1 is',centroids[1])
```

```python
#for iter in range(3):
for i in range(2):
    k=2
    cluster_0=[]
    cluster_1=[]
    for x in data:
        distances=[]
        for c in centroids:
            d=np.dot(x-c,x-c)
            distances.append(d)
        #distances=np.array(distances)
        print('distances are',distances)
        min_dis_index=distances.index(min(distances))
        if(min_dis_index==0):
            cluster_0.append(x)
        else:
            cluster_1.append(x)
            
    cluster_0=np.array(cluster_0)
    print(' updated first cluster are',cluster_0)
    cluster_1=np.array(cluster_1)
    print(' updated second cluster are',cluster_1)
    centroids[0]=cluster_0.mean(axis=0)
    print(' updatd first centroids are',centroids[0])
    centroids[1]=cluster_1.mean(axis=0)
    print(' updatd second centroids are',centroids[1])


print('centroids are',centroids)
    
    
    
    
```

```python
plt.scatter(data[:,0],data[:,1])
plt.plot(centroids[:,0],centroids[:,1],'r*',markersize=20)
plt.show()
```
